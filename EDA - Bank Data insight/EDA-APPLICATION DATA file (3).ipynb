# Outlier Treatment

While analysing the box plot of all columns we can see that IQR and max values are in the particular range and some of the value that lies in a data series on its extremes, which is either very small or large data are considered as outliers

Using mode outliers are treated with the help of upper and lower bound values.

```python
mod_ad.describe()
```

# Capping and Flooring treatment

#### For treating the outliers in series of dataframes, upper bound outliers(>max) are treated by filling the upper bound values and for lower bound outliers are filled with lower bound values as per below formulaes.

```python
for i in Continous_c:
    q1=mod_ad[i].describe()["25%"]
    q3=mod_ad[i].describe()["75%"]
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    mod_ad[i]=np.where(mod_ad[i]>upper_bound,upper_bound,mod_ad[i])## capping
    mod_ad[i]=np.where(mod_ad[i]<lower_bound,lower_bound,mod_ad[i])## flooring
```

# Box plot after outliers treatment
